```python
import pandas as pd
import os
import glob

path = "E:/Data Science/Tasks/AReM"

# Find directories inside path

dir_list = next(os.walk(path))[1]

# Loop to traverse through each directory

for directory in dir_list :
    combined_df = pd.DataFrame()
    print("printing from ",directory)

# Traverse through each .csv file in a directory and append one on top of another

    for file_name in glob.glob(path + "/" + directory + "/*.csv") :
        df = pd.read_csv(file_name,header=None,sep='\t',index_col=False)
        combined_df = combined_df.append(df)
    print(combined_df)

# create a new csv file containing all the concatenated csv files in a folder

    combined_df.to_csv(path + "/" + directory + "/combined.csv",index=False,sep='\t')    
```
